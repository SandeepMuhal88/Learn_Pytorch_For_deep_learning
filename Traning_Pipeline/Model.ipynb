### Defining the model

```python
# Make a class for the model
import torch.nn as nn
import torch.nn.functional as F
class MysimpaleNN():
    def __init__(self,X):
        self.weights=torch.rand(X.shape[1],1,requires_grad=True,dtype=torch.float64)
        self.bias=torch.zeros(1,requires_grad=True,dtype=torch.float64)

    def forward(self,X):
        z=torch.matmul(X,self.weights)+self.bias
        y_pred=torch.sigmoid(z)
        return y_pred
    
    def loss_function(self, y_pred, y):
        # Clamp predictions to avoid log(0)
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

        # Calculate loss
        loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
        return loss
        # loss =torch.nn.BCELoss()
        # return  loss(y_pred, y)
```

### Important Parameters

```python
learning_rate = 0.1
epochs = 25
```

### Training Pipeline

```python
#create an instance of the model
model=MysimpaleNN(X_train_tensor)

# Define loop for training

for i in range(epochs):
    #Foraward pass
    y_pred=model.forward(X_train_tensor)
    # print(y_pred.shape)
    torch.Size([455, 1])
    #Calculate the loss
    loss=model.loss_function(y_pred,y_train_tensor)
    # print(loss)
    # print(loss.shape)

    # Backward pass
    loss.backward()
    # Update parameters
    with torch.no_grad():
        model.weights -= learning_rate * model.weights.grad
        model.bias -= learning_rate * model.bias.grad   
    # Zero gradients
    model.weights.grad.zero_()
    model.bias.grad.zero_()
    # Print loss every epoch
    print(f'Epoch {i+1}/{epochs}, Loss: {loss.item()}')
```
