```python
import torch
import torch.nn as nn
```

```python
class Model(nn.Module):
    def __init__(self,input_features):
    
        super().__init__()
        self.hidden= nn.Linear(input_features, 4)
        self.Relu1= nn.ReLU()
        self.hidden2= nn.Linear(4,3)
        self.Relu2= nn.ReLU()
        self.hidden3= nn.Linear(3,2)
        self.Relu3= nn.ReLU()
        self.hidden4= nn.Linear(2,1)
        self.sigmoid= nn.Sigmoid()

    def forward(self, x):
        out = self.hidden(x)
        out = self.Relu1(out)
        out = self.hidden2(out)
        out = self.Relu2(out)
        out = self.hidden3(out)
        out = self.Relu3(out)
        out = self.hidden4(out)
        out = self.sigmoid(out)
        return x
```

```python
# Create a Dataser for testing

features = torch.rand(100,5)

# Create an instance of the model
model = Model(features.shape[1])

```

```python
model(features)
```

```python
model.hidden.weight
```

```python
from torchinfo import summary
summary(model, input_size=(100, 5), col_names=["input_size", "output_size", "num_params", "trainable"], col_width=20, row_settings=["var_names"])
```

```python
#in this layer we use sequential layers


class Model2(nn.Module):
    def __init__(self, input_features):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(input_features, 4),
            nn.ReLU(),
            nn.Linear(4, 3),
            nn.ReLU(),
            nn.Linear(3, 2),
            nn.ReLU(),
            nn.Linear(2, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        out = self.network(x)
        return out
```

```python
# creat Dataset for testing

features2 = torch.rand(50000, 5)

# Create an instance of the model
model2 = Model2(features2.shape[1])
model2(features2)
```

```python
from torchinfo import summary
summary(model2, input_size=(100, 5), col_names=["input_size", "output_size", "num_params", "trainable"], col_width=20, row_settings=["var_names"])
```
