```python
import pandas as pd
import numpy as np 
import torch
import torch.nn as nn
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("fedesoriano/cifar10-python-in-csv")

print("Path to dataset files:", path)
```

```python
import os 

print("Dictionary:- ",os.listdir(path))
```

```python
#  1. Create an empty list to hold the individual DataFrames
list_of_dataframes = []

# 2. Loop through each file in the directory
for filename in os.listdir(path):
    # 3. Check if the file is a CSV
    if filename.endswith('.csv'):
        # 4. Construct the full path to the file
        # os.listdir() only gives the filename, not the full path
        full_path = os.path.join(path, filename)
        
        # 5. Read the CSV into a DataFrame and append it to your list
        df_temp = pd.read_csv(full_path)
        list_of_dataframes.append(df_temp)
```

```python
test_df=list_of_dataframes[0]
train_df=list_of_dataframes[1]


test_df.head()
```
